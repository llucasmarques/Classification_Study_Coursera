```python
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score

diabetes = pd.read_csv ('diabetes.csv')
diabetes.head()
```

In this dataset, the data was extracted of many pacients that realized the Diabetes Test.

This Dataset is composed of 768 pacients wich include 500 pacients tested as Negative and 268 Positives.

Objective:
Classify the diagnosis in two categories:
Positive or Negative

```python
#Correlation Heatmap
plt.figure(figsize=(16, 6))
heatmap = sn.heatmap(diabetes.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
x = diabetes.drop('resultado', axis=1) #features
y = diabetes['resultado'] #label
x_train , x_test, y_train, y_test = train_test_split(x, y) #Spliting my dataset (test and training) 
```

```python
# Using Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier

SEED = 4121988
clf_tree = DecisionTreeClassifier(random_state=SEED, max_depth=3)
clf_tree.fit(x_train, y_train)
clf_tree_y_predict = clf_tree.predict(x_test)

# Using SVM
clf_svc = svm.SVC().fit(x_train, y_train)
clf_svc_y_predict = clf_svc.predict(x_test)

# Using Random Forest
clf_randomForest = RandomForestClassifier()
clf_randomForest.fit(x_train, y_train)
clf_randomForest_y_predict = clf_randomForest.predict(x_test)
```

```python
#Print SVM Model Results
print('Accuracy train: ', accuracy_score(y_train, clf_svc.predict(x_train)), '/ Accuracy test: ', accuracy_score(y_test, clf_svc_y_predict) )
print('Precision Score train: ', precision_score(y_train, clf_svc.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Precision Score test: ', precision_score(y_test, clf_svc_y_predict,  average="binary", pos_label="testado_positivo"))
print('Recall Score train: ', recall_score(y_train, clf_svc.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Recall Score test: ',  recall_score(y_test, clf_svc_y_predict, average="binary", pos_label="testado_positivo") )
print('F1 Score train: ', f1_score(y_train, clf_svc.predict(x_train), average="binary", pos_label="testado_positivo"), '/ F1 Score test: ',  f1_score(y_test, clf_svc_y_predict, average="binary", pos_label="testado_positivo") )
```

```python
#Print Decision Tree Model Results
print('Accuracy train: ', accuracy_score(y_train, clf_tree.predict(x_train)), '/ Accuracy test: ', accuracy_score(y_test, clf_tree_y_predict) )
print('Precision Score train: ', precision_score(y_train, clf_tree.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Precision Score test: ', precision_score(y_test, clf_tree_y_predict,  average="binary", pos_label="testado_positivo"))
print('Recall Score train: ', recall_score(y_train, clf_tree.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Recall Score test: ',  recall_score(y_test, clf_tree_y_predict, average="binary", pos_label="testado_positivo") )
print('F1 Score train: ', f1_score(y_train, clf_tree.predict(x_train), average="binary", pos_label="testado_positivo"), '/ F1 Score test: ',  f1_score(y_test, clf_tree_y_predict, average="binary", pos_label="testado_positivo") )
```

```python
#Print RandomForest Model Results
print('Accuracy train: ', accuracy_score(y_train, clf_randomForest.predict(x_train)), '/ Accuracy test: ', accuracy_score(y_test, clf_randomForest_y_predict) )
print('Precision Score train: ', precision_score(y_train, clf_randomForest.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Precision Score test: ', precision_score(y_test, clf_randomForest_y_predict,  average="binary", pos_label="testado_positivo"))
print('Recall Score train: ', recall_score(y_train, clf_randomForest.predict(x_train), average="binary", pos_label="testado_positivo"), '/ Recall Score test: ',  recall_score(y_test, clf_randomForest_y_predict, average="binary", pos_label="testado_positivo") )
print('F1 Score train: ', f1_score(y_train, clf_randomForest.predict(x_train), average="binary", pos_label="testado_positivo"), '/ F1 Score test: ',  f1_score(y_test, clf_randomForest_y_predict, average="binary", pos_label="testado_positivo") )
```

The Random Forest provided best Scores comparing with DecisionTree and SVM model.

In this case, I will continue with Random Forest.
